# Objective 2 : Data Cleaning and Preprocessing

```python
# Cleaning by checking on missing values, dropping any NA values
print(df.dropna()) # conclusion - dont have any NA values
```

```python
# remove duplicates
print(df.drop_duplicates(inplace=True)) 
display(df)
```

```python
# A total of 62 duplicates were removed
```

```python
# Data type conversion:
df['Income'] = df['Income'].astype(float)

# Feature creation:

# Creating Score_Numeric as an additional column for coorelation 
score_map = {'Low': 0, 'Average': 1, 'High': 2}
df['Score_Numeric'] = df['Credit Score'].map(score_map)

# Creating Risk_Category as an additional column for Risk bucketization
risk_map = {'Low': 'High Risk', 'Average': 'Medium Risk', 'High': 'Low Risk'}
df['Risk_Category'] = df['Credit Score'].map(risk_map) 

display(df.head())
```

# Objective 4: Statistical and Correlation Analysis

```python
# Statistical analysis on columns
display(df.describe())  # showing statistical values for columns with integer and float values
display(df.describe(include='all')) # show statistical valuees for the entire dataset 
display(df.describe(include=['int64'])) # show statistical values for columns having integer values
display(df.describe(include=['object'])) # show statistical values for columns having string values
```

```python
# Correlation between columns that impacts credit score
print(df[['Age', 'Income', 'Number of Children', 'Score_Numeric']].corr())
```

```python
# Statistical analysis using group by with numerical columns
display(df.groupby('Credit Score')['Income'].median())
display(df.groupby('Credit Score')[['Age','Income']].mean())
```

# Objective 5: Visualization and Trend Analysis

```python
# 1. Countplot to display the frequency of each credit score category in the dataset for 
# the Credit Score vs. Risk Category plot, along with risk interpretation.
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Credit Score', hue='Risk_Category', palette=['Green','Orange','Red'])
plt.title('Credit Score vs Risk Category')
plt.show()

display(df['Risk_Category'].value_counts().reset_index())

# 2. Box Plot: Income vs Credit Score
# To understand distribution of Income based on Credit score - High Income is proportional to high credit score and like wise
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Credit Score', y='Income')
plt.title('Income Distribution by Credit Score')
plt.show()

# 3. Heatmap for Correlation
# Correlation between numerical columns which impacts credit score
cols = ['Age', 'Income', 'Number of Children', 'Score_Numeric']
corr_matrix = df[cols].corr()
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
